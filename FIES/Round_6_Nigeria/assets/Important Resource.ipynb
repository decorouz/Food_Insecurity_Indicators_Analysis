### Prevalence of Food Insecurity with FIES [app](https://fies.shinyapps.io/ExtendedApp/):
To obtain internationally comparable estimates of the prevalence of FI (`FImod+sev and FIsev`) suggested by the FAO and by utilizing the Voices of the Hungry project’s FIES app.


Two globally-comparable indicators of the prevalence of FI at different levels of severity: **moderate or severe FI** (raw score = 5 to 7) and **severe FI** (raw score = 8), were calculated. All analyses were adjusted for sampling weights.

To receive comparable estimates of the prevalence of FI, we inserted uploaded the original FIES data to the Voices of the Hungry project’s FIES app. We examined the graphical representation of the item parameters. 

The item `fies_skipped` was the furthest away from the diagonal line, with the highest absolute difference compared to all other items(Fig 1)

The correlation between item parameters of the two scales was 90.1 %. After excluding the item `feis_skipped` the correlation between item parameters increased to 94.9 %. Due to a high absolute difference between the country and global standard items of `fies_ateless` (Fig. 2), we decided to omit `fies_ateless` which led to an increase in the correlation between the common items to 98.2 %. We choose this as best equating scenario (Fig. 3). 


The resulting proportion of the sample population experiencing moderate or severe FI (`FImod+sev`) is 78.71 %. The proportion of the population experiencing severe FI (`FIsev`) is 17.13 %.

```python
import math
import numpy as np
from scipy.stats import lognorm
import matplotlib.pyplot as plt
import statsmodels.api as sm

#make this example reproducible
np.random.seed(1)

#generate dataset that contains 1000 log-normal distributed values
lognorm_dataset = lognorm.rvs(s=.5, scale=math.exp(1), size=1000)

#create histogram to visualize values in dataset
plt.hist(lognorm_dataset, edgecolor='black', bins=20)
```

```python
#create Q-Q plot with 45-degree line added to plot
fig = sm.qqplot(lognorm_dataset, line='45')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating DataFrames
data = {
    'food groups': ['Grains and Tubers', 'Meat and Poultry', 'Other VitA Veg', 'Pulses', 'Nuts and Seeds',
                    'Diary', 'Eggs', 'Dark Green Veg', 'Other Veg', 'Other Fruits', 'Legumes and Nuts',
                    'Breast Milk', 'Other Fruits and Veg'],
    'Women': [98.7, 18.5, 6.3, 5.4, 23.2, 8.9, 0.4, 69.2, 84.4, 14.8, 0, 0, 0],
    'Child': [97, 24.4, 10.3, 0, 0, 15, 3.4, 0, 0, 0, 65.4, 47.01, 87.6]
}

df = pd.DataFrame(data)

data2 = {
    'Women': [77.5, 22.5, 0, 0],
    'Child': [0, 0, 76, 24]
}

# Define the index labels
index_labels = ['< 5 Food Groups', '≥ 5 Food Groups', '< 4 Food Groups', '≥ 4 Food Groups']

# Creating the second DataFrame
df1 = pd.DataFrame(data2, index=index_labels)

from matplotlib import rc

from matplotlib import rcParams

#rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('font',**{'family':'serif','serif':['Times']})

# Create subplots with shared y-axis
fig, axes = plt.subplots(1, 2, figsize=(20, 10), sharey=True)  # Set sharey=True
# Plot for df2
df1.plot(kind='bar', ax=axes[0])
# axes[0].set_title('Food Groups Consumption by Category')
# axes[0].set_xlabel('Category')
axes[0].set_ylabel('Percentage (%)', fontsize=14)
axes[0].legend( loc='upper center', ncols=2, fontsize=14,labelcolor='black')
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)
axes[0].spines['bottom'].set_visible(False)
axes[0].spines['left'].set_visible(False)
axes[0].tick_params(axis='x', which='both', bottom=False, top=False)  # Remove x-axis ticks
axes[0].tick_params(axis='y', which='both', left=False, right=False)  # Remove y-axis ticks
axes[0].grid(axis='y', linestyle='--', alpha=0.5)  # Add faint grid lines


COLOR = 'black'
plt.rcParams['xtick.color'] = COLOR

# Plot for df1
df.set_index('food groups').plot(kind='bar', ax=axes[1],width=0.8 )
# axes[1].set_title('Food Groups Consumption by Group')
axes[1].set_xlabel("")
axes[1].set_ylabel('Percentage')
axes[1].legend("")

axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)
axes[1].spines['bottom'].set_visible(False)
axes[1].spines['left'].set_visible(False)
axes[1].tick_params(axis='x', which='both', bottom=False, top=False)  # Remove x-axis ticks
axes[1].tick_params(axis='y', which='both', left=False, right=False)  # Remove y-axis ticks
axes[1].grid(axis='y', linestyle='--', alpha=0.4)  # Add faint grid lines

xlabels = [label.replace(' ', '\n') if len(label) > 4 else label for label in df['food groups']]
# xlabels2 = [label.replace(' ', '\n') if len(label) > 4 else label for label in index_labels]

axes[0].set_xticks(ticks=range(len(index_labels)), labels=index_labels, rotation=0, ha='center', fontsize=14)

axes[1].set_xticks(ticks=range(len(df['food groups'])), labels=xlabels, rotation=0, ha='center', fontsize=14)

plt.tight_layout()
plt.show()
```
